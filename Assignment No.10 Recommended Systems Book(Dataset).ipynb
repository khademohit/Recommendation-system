```python
book2 = book1.drop(['Unnamed: 0'],axis=1)
```

```python
book2
```

```python
book2 = book2.sort_values(['UserId'])
```

```python
book3 = book2.drop_duplicates()
book3
```

```python
book4 = book3.pivot_table(index='UserId',columns='Title',values='Rating').reset_index()
```

```python
book4
```

```python
book4.fillna(0, inplace=True)
```

```python
book4
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1 - pairwise_distances( book4.values,metric='cosine')
```

```python
user_sim
```

```python
book_df = pd.DataFrame(user_sim)
```

```python
book_df.index = book4.UserId.unique()
book_df.columns = book4.UserId.unique()
```

```python
book_df.iloc[0:10,0:10]
```

```python
np.fill_diagonal(user_sim, 0)
book_df.iloc[0:10, 0:10]
```
